```java
interface Meal {
    void prepare();
}

class Burger implements Meal {
    public void prepare() {
        System.out.println("Preparing Burger Meal");
    }
}

class Pizza implements Meal {
    public void prepare() {
        System.out.println("Preparing Pizza Meal");
    }
}

class Salad implements Meal {
    public void prepare() {
        System.out.println("Preparing Salad Meal");
    }
}

class MealFactory {
    public static Meal createMeal(String type) {
        if (type.equalsIgnoreCase("burger")) {
            return new Burger();
        } else if (type.equalsIgnoreCase("pizza")) {
            return new Pizza();
        } else if (type.equalsIgnoreCase("salad")) {
            return new Salad();
        }

        return null;
    }
}
```

```java
class GameSettings {
    private static GameSettings instance;

    private String difficulty;

    private GameSettings() {
        difficulty = "Normal";
    }

    public static GameSettings getInstance() {
        if (instance == null) {
            instance = new GameSettings();
        }

        return instance;
    }

    public void setDifficulty(String difficulty) {
        this.difficulty = difficulty;
    }

    public String getDifficulty() {
        return difficulty;
    }
}
```

```java
Meal burger = MealFactory.createMeal("burger");
Meal pizza = MealFactory.createMeal("pizza");
Meal salad = MealFactory.createMeal("salad");
Meal invalid = MealFactory.createMeal("pasta");

assert burger instanceof Burger : "Burger creation failed";
assert pizza instanceof Pizza : "Pizza creation failed";
assert salad instanceof Salad : "Salad creation failed";
assert invalid == null : "Invalid meal should return null";

System.out.println("Static Factory assertions passed");
```

```java
GameSettings settings1 = GameSettings.getInstance();
GameSettings settings2 = GameSettings.getInstance();

assert settings1 == settings2 : "Singleton instances are not the same";

settings1.setDifficulty("Hard");

assert settings2.getDifficulty().equals("Hard") : "Singleton state not shared";

System.out.println("Singleton assertions passed");
```
